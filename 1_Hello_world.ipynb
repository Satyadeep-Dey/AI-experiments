```python
text = "Hello World"
print(len(text)) # calculates length of string
print(text.replace("World","Planet")) #replace one string with another
text ="     This sentence has white spaces   "
print(text)
print("***********Length before strip is - " + str(len(text)) + "  ***********")
print(text.strip())
print("***********Length after strip is - "+ str(len(text.strip())) + "  ***********")
text = "Hello everyone, Sachin is great"
words = text.split()
# Divides a string into a list of substrings based on a specified separator (default is whitespace).
print(words)
sentence = " ".join(words)
#Combines a list of strings into a single string, with each element separated by the string it was called on.
print(sentence)
words = text.split(",") # split on comma , not whitespace
print(words)
print(words[0])
###################
multi_line_string = "This is a string that is too long to fit on one line, so we continue it \
here without any breaks."
#This prints as a single line, with no breaks, because the backslash \ tells Python to ignore the line break.
print(multi_line_string)
###################
multi_line_quote = """This is a string
that contains multiple lines,
and we don't need to use a backslash.
It also handles quotes, like "this one" and 'that one'."""
#Triple quotes allow for direct line breaks, and you can include both single (') and double (") quotes.
print(multi_line_quote)
```
