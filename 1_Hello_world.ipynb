GENERATORS
1. The yield keyword is the core of a generator. When a function has yield instead of return, it becomes a generator function. Each time yield is called, it pauses the function, saving its state, and returns a value to the caller.
2. Memory Efficiency: They don’t store all values in memory; they generate each value only when needed.
3. Lazy Evaluation: Generators don’t calculate values until you actually need them, which saves memory.
4. Simplifies Code: They’re a great way to handle sequences that are calculated on-the-fly without needing complex data structures.

```python
def number_generator():
    for i in range(1,6):
        yield i
```

```python
# Create a generator
gen = number_generator()

# Use a loop to get values from the generator
for number in gen:
    print(number)
```

Explanation of the Output

1. When the for loop starts, gen begins at the first value (1).
2. It calls yield and outputs 1, then pauses.
3. Each time the loop continues, the generator resumes from where it left off, producing 2, 3, 4, and finally 5.
4. After yielding 5, the generator ends.

```python
#Alternatively, you can call next() on the generator to get values one by one.
gen = number_generator()
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
# Keep calling next until you get a StopIteration error
```

```python
#dictionary
book1 = {"title": "Great Expectations", "author": "Charles Dickens"}
book2 = {"title": "Bleak House", "author": "Charles Dickens"}
book3 = {"title": "A Book By No Author"}
book4 = {"title": "Moby Dick", "author": "Herman Melville"}
#list
books = [book1, book2, book3, book4]

```

```python
[book.get('author') for book in books ] # dictionary
```

```python
[book.get('author') for book in books if book.get('author')] # dictionary .. we filter out books with no author
```

```python
{book.get('author') for book in books if book.get('author')} # set and hence unique
```

```python
# You can use "yield from" to yield each item of an iterable

def authors_generator():
    yield from {book.get("author") for book in books if book.get("author")}
```

```python
# Use the generator

for author in authors_generator():
    print(author)
```
